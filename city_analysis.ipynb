```python
import seaborn as sns  # visualization tool
```

```python
yongin = df.loc[df['시군구명'].str.startswith('용인시')]
anyang = df.loc[df['시군구명'].str.startswith('안양시')]
```

```python
anyang.dtypes
```

```python
print(len(set(df['상권업종대분류명'])))
print(len(set(df['상권업종중분류명'])))
print(len(set(df['상권업종소분류명'])))
print(len(set(df['표준산업분류명'])))
```

```python
fig = plt.figure(figsize=(8, 8))
sns.countplot(x='상권업종대분류명', data=yongin, order = yongin['상권업종대분류명'].value_counts().index)
```

```python
fig = plt.figure(figsize=(8, 8))
sns.countplot(x='상권업종대분류명', data=anyang, order = anyang['상권업종대분류명'].value_counts().index)
```

```python
fig = plt.figure(figsize=(16, 8))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

ax1 = fig.add_subplot(2, 1, 1)
yongin_food = yongin.loc[yongin['상권업종대분류명'] == '음식']
sns.countplot(x='상권업종중분류명', data=yongin_food, order = yongin_food['상권업종중분류명'].value_counts().index, ax=ax1)

ax2 = fig.add_subplot(2, 1, 2)
anyang_health = anyang.loc[anyang['상권업종대분류명'] == '의료']
sns.countplot(x='상권업종중분류명', data=anyang_health, order = anyang_health['상권업종중분류명'].value_counts().index, ax=ax2)
```

```python
fig = plt.figure(figsize=(24, 40))
fig.subplots_adjust(hspace=0.4, wspace=0.4)
n = len(set(df['상권업종대분류명']))
t = 1
for i in ['관광/여가/오락', '부동산', '생활서비스', '소매', '숙박', '스포츠', '음식', '의료', '학문/교육']:#set(df['상권업종대분류명']):
    ax1 = fig.add_subplot(n, 1, t)
    t= t+ 1
    yongin_x = yongin.loc[yongin['상권업종대분류명'] == i]
    sns.countplot(x='상권업종중분류명', data=yongin_x, order = yongin_x['상권업종중분류명'].value_counts().index, ax=ax1)
```
